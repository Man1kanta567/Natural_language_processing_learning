```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import  re
```

Data Extraction

```python
df = pd.read_csv(filepath_or_buffer='SMSSpamCollection.txt',sep='\t',names=['label','messaage'])

df.head(5)
```

Data cleaning and preprocessing


preprocessing include the removal of unneccesary words in vocabulary based ont he stop words

```python
ps = PorterStemmer()

corpus:list = []
for i in range(0,len(df)):
    sentence = re.sub('[^a-zA-Z]', ' ', df.iloc[i]['messaage'])
    words :list = sentence.split()
    words = [ps.stem(word=word) for word in words if word not in stopwords.words('english') ] 
    sentence =' '.join(words)
    corpus.append(sentence)
    
    
```

```python
Y = df['label'].values
```

```python
Y = pd.get_dummies(df['label'])
```

```python
y=Y.iloc[:,0].values
```

```python
y
```

```python
import numpy as np
y=np.where(y,1,0)
print(y)
```

Spilt the train and test data 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(corpus,y,test_size=0.2)
```

```python
Y_train
```

```python
X_train
```

Apply BagOfWords and fetch the features


Covert the data into metrix 

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=1000,binary=True,ngram_range=(2,2))
X=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()
```

```python
X
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import  MultinomialNB
mnb = MultinomialNB()
mnb.fit(X=X,y=Y_train)
```

```python
Y_pred = mnb.predict(X_test)
```

check the peformance and accuracy of the model

```python
from sklearn.metrics import accuracy_score,classification_report

print('accuracy_score : ',accuracy_score(y_true=Y_test,y_pred=Y_pred))
print('classification report :',classification_report(y_true=Y_test,y_pred=Y_pred))
```
