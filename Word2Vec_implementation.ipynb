```python
import numpy as np
import pandas as pd
import nltk
import gensim

```

```python
df = pd.read_csv(filepath_or_buffer='SMSSpamCollection.txt',sep='\t',names=['label','message'])
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words:list=[]
for i in range(0,len(df)):
    setence = sent_tokenize(text=df['message'][i])
    for sent in setence:
        words.append(simple_preprocess(sent))
        
```

```python
words
```

```python
from gensim.models import Word2Vec
model =Word2Vec(sentences=words,epochs=10)
```

```python
def avg_word2vec(doc):
    return np.mean([model.wv[w] for w in doc if w in model.wv], axis=0)
```

```python
import tqdm
```

```python
X =[]
for i in tqdm.tqdm(range(len(df))):
    X.append(avg_word2vec(words[i]))
```

```python
np.array(X[0]).shape
```

```python
X_clean = [x for x in X if isinstance(x, np.ndarray)]
X_clean=np.array(X_clean)
```
